```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pickle
```

```python
traits=['EI','WI','EC','WC','EA','WA','E','A','N','O']
```

```python
def plot_curves(trait):
    with open(f'trait-wise/without_reliability/{trait}_accuracy.pkl','rb') as f:
        without_rel_acc=np.array(pickle.load(f))
    with open(f'trait-wise/without_reliability/{trait}_confidence.pkl','rb') as f:
        without_rel_con=np.array(pickle.load(f))
    with open(f'trait-wise/with_reliability/reliability_{trait}_confidence.pkl','rb') as f:
        with_rel_con=np.array(pickle.load(f))
    with open(f'trait-wise/with_reliability/reliability_{trait}_accuracy.pkl','rb') as f:
        with_rel_acc=np.array(pickle.load(f))
    with open(f'trait-wise/corrected/{trait}_accuracy.pkl','rb') as f:
        corr_acc=np.array(pickle.load(f))
    with open(f'trait-wise/corrected/{trait}_confidence.pkl','rb') as f:
        corr_con=np.array(pickle.load(f))
    with open(f'trait-wise/augmented/{trait}_accuracy.pkl','rb') as f:
        aug_acc=np.array(pickle.load(f))
    with open(f'trait-wise/augmented/{trait}_confidence.pkl','rb') as f:
        aug_con=np.array(pickle.load(f))
    l=len(aug_acc)
    x=[i+1 for i in range(l)]
    models=['Logistic Regression','SVM','Neural Network','Random Forest']
    for i in range(4):
        fig,axes=plt.subplots(1,2,figsize=(10,5))
        ax1=axes[0]
        ax2=axes[1]
        ax1.plot(x,without_rel_acc[:,i],color='r',label='Normal')
        ax1.plot(x,with_rel_acc[:,i],color='g',label='Concatenating reliability vector')
        ax1.plot(x,corr_acc[:,i],color='b',label='Replacing with average response')
        ax1.plot(x,aug_acc[:,i],color='yellow',label='Using augmented dataset')
        ax1.set_xlabel('No. of questions',fontsize='8')
        ax1.set_ylabel('Accuracy',fontsize='8')
        ax2.plot(x,without_rel_con[:,i],color='r',label='Normal')
        ax2.plot(x,with_rel_con[:,i],color='g',label='Concatenating reliability vector')
        ax2.plot(x,corr_con[:,i],color='b',label='Replacing with average response')
        ax2.plot(x,aug_con[:,i],color='yellow',label='Using augmented dataset')
        ax2.set_xlabel('No. of questions',fontsize='8')
        ax2.set_ylabel('Confidence',fontsize='8')
        fig.suptitle(models[i])
        plt.legend(fontsize='8')
        plt.show()
```

```python
for trait in traits:
    print(trait)
    plot_curves(trait)
```
