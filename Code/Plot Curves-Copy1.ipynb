```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pickle
```

```python
def plot_curves():
    with open('combined_accuracy.pkl','rb') as f:
        acc=np.array(pickle.load(f))
    with open('combined_confidence.pkl','rb') as f:
        con=np.array(pickle.load(f))
    with open('corrected_combined_accuracy.pkl','rb') as f:
        corr_acc=np.array(pickle.load(f))
    with open('corrected_combined_confidence.pkl','rb') as f:
        corr_con=np.array(pickle.load(f))
    with open('augmented_combined_accuracy.pkl','rb') as f:
        aug_acc=np.array(pickle.load(f))
    with open('augmented_combined_confidence.pkl','rb') as f:
        aug_con=np.array(pickle.load(f))
    with open('reliability_combined_accuracy.pkl','rb') as f:
        rel_acc=np.array(pickle.load(f))
    with open('reliability_combined_confidence.pkl','rb') as f:
        rel_con=np.array(pickle.load(f))
    with open('bert_accuracy.pkl','rb') as f:
        bert_acc=np.array(pickle.load(f))
    with open('bert_confidence.pkl','rb') as f:
        bert_con=np.array(pickle.load(f))
    with open('glove_accuracy.pkl','rb') as f:
        glove_acc=np.array(pickle.load(f))
    with open('glove_confidence.pkl','rb') as f:
        glove_con=np.array(pickle.load(f))
    with open('use_accuracy.pkl','rb') as f:
        use_acc=np.array(pickle.load(f))
    with open('use_confidence.pkl','rb') as f:
        use_con=np.array(pickle.load(f))
    l=len(acc)
    x=[i+1 for i in range(l)]
    models=['Logistic Regression','SVM','Neural Network','Random Forest']
    for i in range(4):
        fig,axes=plt.subplots(1,2,figsize=(10,5))
        ax1=axes[0]
        ax2=axes[1]
        ax1.plot(x,acc[:,i],color='r',label='Normal')
        ax1.plot(x,rel_acc[:,i],color='orange',label='Concatenating reliability vector')
        ax1.plot(x,corr_acc[:,i],color='b',label='Replacing with average response')
        ax1.plot(x,aug_acc[:,i],color='yellow',label='Using augmented dataset')
        ax1.plot(x,bert_acc[:,i],color='g',label='Using BERT for augmenting')
        ax1.plot(x,glove_acc[:,i],color='m',label='Using Glove for augmenting')
        ax1.plot(x,use_acc[:,i],color='cyan',label='Using Universal Sentence Encoder')
        ax1.set_xlabel('No. of questions',fontsize='8')
        ax1.set_ylabel('Accuracy',fontsize='8')
        ax2.plot(x,con[:,i],color='r',label='Normal')
        ax2.plot(x,rel_con[:,i],color='orange',label='Concatenating reliability vector')
        ax2.plot(x,corr_con[:,i],color='b',label='Replacing with average response')
        ax2.plot(x,aug_con[:,i],color='yellow',label='Using augmented dataset')
        ax2.plot(x,bert_con[:,i],color='g',label='Using BERT for augmenting')
        ax2.plot(x,glove_con[:,i],color='m',label='Using Glove for augmenting')
        ax2.plot(x,use_con[:,i],color='cyan',label='Using Universal Sentence Encoder')
        ax2.set_xlabel('No. of questions',fontsize='8')
        ax2.set_ylabel('Confidence',fontsize='8')
        fig.suptitle(models[i])
        plt.legend(fontsize='8')
        plt.show()
    acc1=[]
    acc2=[]
    acc3=[]
    acc4=[]
    acc5=[]
    acc6=[]
    acc7=[]
    for i in range(l):
        acc1.append(max(acc[i]))
        acc2.append(max(rel_acc[i]))
        acc3.append(max(corr_acc[i]))
        acc4.append(max(aug_acc[i]))
        acc5.append(max(bert_acc[i]))
        acc6.append(max(glove_acc[i]))
        acc7.append(max(use_acc[i]))
    plt.plot(x,acc1,color='r',label='Normal')
    plt.plot(x,acc2,color='orange',label='Concatenating reliability vector')
    plt.plot(x,acc3,color='b',label='Replacing with average response')
    plt.plot(x,acc4,color='yellow',label='Using augmented dataset')
    plt.plot(x,acc5,color='g',label='Using BERT for augmenting')
    plt.plot(x,acc6,color='m',label='Using Glove for augmenting')
    plt.plot(x,acc7,color='cyan',label='Using Universal Sentence Encoder')
    plt.xlabel('No. of questions',fontsize='8')
    plt.ylabel('Accuracy',fontsize='8')
    plt.title('Max accuracy plot')
    plt.legend(fontsize='8')
    plt.show()
```

```python
plot_curves()
```

```python
def plot2():
    with open('bert_accuracy.pkl','rb') as f:
        bert_acc=np.array(pickle.load(f))
    with open('bert_confidence.pkl','rb') as f:
        bert_con=np.array(pickle.load(f))
    with open('true_bert_accuracy.pkl','rb') as f:
        bert2_acc=np.array(pickle.load(f))
    with open('true_bert_confidence.pkl','rb') as f:
        bert2_con=np.array(pickle.load(f))
    l=len(bert_acc)
    x=[i+1 for i in range(l)]
    models=['Logistic Regression','SVM','Neural Network','Random Forest']
    for i in range(4):
        fig,axes=plt.subplots(1,2,figsize=(10,5))
        ax1=axes[0]
        ax2=axes[1]
        ax1.plot(x,bert_acc[:,i],color='g',label='Augmented test-set')
        ax1.plot(x,bert2_acc[:,i],color='r',label='Usual test-set')
        ax1.set_xlabel('No. of questions',fontsize='8')
        ax1.set_ylabel('Accuracy',fontsize='8')
        ax2.plot(x,bert_con[:,i],color='g',label='Augmented test-set')
        ax2.plot(x,bert2_con[:,i],color='r',label='Usual test-set')
        ax2.set_xlabel('No. of questions',fontsize='8')
        ax2.set_ylabel('Confidence',fontsize='8')
        fig.suptitle(models[i])
        plt.legend(fontsize='8')
        plt.show()
    acc1=[]
    acc2=[]
    for i in range(l):
        acc1.append(max(bert_acc[i]))
        acc2.append(max(bert2_acc[i]))
    plt.plot(x,acc1,color='g',label='Augmented test-set')
    plt.plot(x,acc2,color='r',label='Usual test-set')
    plt.xlabel('No. of questions',fontsize='8')
    plt.ylabel('Accuracy',fontsize='8')
    plt.title('Max accuracy plot')
    plt.legend(fontsize='8')
    plt.show()
```

```python
plot2()
```
